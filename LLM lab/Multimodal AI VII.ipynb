```python
# ============================================================
# MULTIMODAL LLM DEMO USING HUGGING FACE TRANSFORMERS
# Includes:
# 1. Image Captioning
# 2. Visual Question Answering (VQA)
# 3. Text Generation
# 4. Ethical Implication Analysis
# ============================================================

# -----------------------------
# STEP 1: Install Dependencies
# -----------------------------
# (Run this in Colab or terminal if needed)
# !pip install transformers torch pillow

# -----------------------------
# STEP 2: Import Libraries
# -----------------------------
import torch
from transformers import pipeline
from PIL import Image
import requests

# -----------------------------
# STEP 3: Load an Image
# -----------------------------
# Using a sample image from the web
image_url = "https://huggingface.co/datasets/Narsil/image_dummy/raw/main/parrots.png"
image = Image.open(requests.get(image_url, stream=True).raw)

# -----------------------------
# STEP 4: Image Captioning Model
# -----------------------------
# This model generates a textual description of the image
caption_pipeline = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")

caption = caption_pipeline(image)[0]['generated_text']
print("\n🖼️ IMAGE CAPTION:")
print(caption)

# -----------------------------
# STEP 5: Visual Question Answering (VQA)
# -----------------------------
# This model answers questions based on image content
vqa_pipeline = pipeline("visual-question-answering", model="dandelin/vilt-b32-finetuned-vqa")

question = "How many birds are in the image?"
vqa_result = vqa_pipeline(image=image, question=question)

print("\n❓ VQA RESULT:")
print(f"Question: {question}")
print(f"Answer: {vqa_result[0]['answer']}")

# -----------------------------
# STEP 6: Text Generation (LLM)
# -----------------------------
# Generates explanation or reasoning
text_generator = pipeline("text-generation", model="gpt2")

prompt = f"Explain the scene: {caption}"
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)

print("\n📝 GENERATED TEXT:")
print(generated_text[0]['generated_text'])

# -----------------------------
# STEP 7: Ethical Implication Analysis
# -----------------------------
def ethical_analysis(caption, generated_text):
    """
    Simple rule-based ethical analysis of outputs
    """

    print("\n⚖️ ETHICAL ANALYSIS REPORT")
    print("="*50)

    # 1. Bias Detection (basic keyword-based check)
    bias_keywords = ["man", "woman", "black", "white", "rich", "poor"]
    bias_found = any(word in caption.lower() for word in bias_keywords)

    print("Bias Check:")
    if bias_found:
        print("⚠️ Potential bias detected in caption.")
    else:
        print("✅ No obvious bias detected.")

    # 2. Hallucination Check
    # If generated text contains words not in caption, possible hallucination
    caption_words = set(caption.lower().split())
    generated_words = set(generated_text.lower().split())

    hallucination = len(generated_words - caption_words) > 10

    print("\nHallucination Check:")
    if hallucination:
        print("⚠️ Possible hallucination detected in generated text.")
    else:
        print("✅ Low hallucination risk.")

    # 3. Privacy Check
    privacy_keywords = ["name", "identity", "address", "person"]
    privacy_issue = any(word in generated_text.lower() for word in privacy_keywords)

    print("\nPrivacy Check:")
    if privacy_issue:
        print("⚠️ Potential privacy concern.")
    else:
        print("✅ No privacy issues detected.")

    print("="*50)


# Run Ethical Analysis
ethical_analysis(caption, generated_text[0]['generated_text'])

# -----------------------------
# STEP 8: Summary
# -----------------------------
print("\n🎯 SUMMARY:")
print("This program demonstrates how multimodal LLMs can:")
print("1. Understand images (captioning)")
print("2. Answer questions from images (VQA)")
print("3. Generate human-like explanations")
print("4. Raise ethical concerns like bias, hallucination, and privacy")
```
