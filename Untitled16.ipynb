```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import Trainer, TrainingArguments
from datasets import load_dataset
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
```

```python
from datasets import load_dataset

dataset = load_dataset("imdb")

train_data = dataset["train"]
test_data = dataset["test"]

# reduce dataset for fast training
train_data = train_data.shuffle(seed=42).select(range(2000))
test_data = test_data.shuffle(seed=42).select(range(500))
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize(example):
    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=256
    )

train_data = train_data.map(tokenize, batched=True)
test_data = test_data.map(tokenize, batched=True)

train_data.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
test_data.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}
```

```python
model_freeze = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
```

```python
model_freeze = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

```python
for param in model_freeze.bert.parameters():
    param.requires_grad = False
```

```python
training_args_freeze = TrainingArguments(
    output_dir="./results_freeze",
    learning_rate=2e-4,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    logging_steps=100
)
```

```python
trainer_freeze = Trainer(
    model=model_freeze,
    args=training_args_freeze,
    train_dataset=train_data,
    eval_dataset=test_data,
    compute_metrics=compute_metrics,
)

trainer_freeze.train()
freeze_results = trainer_freeze.evaluate()

print("Frozen model:", freeze_results)
```

```python
model_ft = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

```python
training_args_ft = TrainingArguments(
    output_dir="./results_finetune",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    logging_steps=100
)
```

```python
trainer_ft = Trainer(
    model=model_ft,
    args=training_args_ft,
    train_dataset=train_data,
    eval_dataset=test_data,
    compute_metrics=compute_metrics,
)

trainer_ft.train()
finetune_results = trainer_ft.evaluate()

print("Fine-tuned model:", finetune_results)
```

```python
print("\nFINAL COMPARISON")
print("Frozen:", freeze_results)
print("Fine-tuned:", finetune_results)
```
